```python
from collections      import deque
from keras.models     import Sequential
from keras.layers     import Dense
from keras.optimizers import Adam
from keras.utils import to_categorical
import numpy as np
import pandas as pd
```

```python
filename = "//222.110.192.241/Work/평가사업본부/외화상품실/00.개인폴더/김남호/Py/lakiaro/price.csv"
df = pd.read_csv(filename, encoding='cp949')
print(len(df))

#전처리
def norm(data):
    lst = []
    for i in range(len(data)):
        if i+1 == len(data):
            break
        else:
            dP = (data[i+1]-data[i])/data[i]
            lst.append(dP)
    lst = np.array(lst)
    return lst

train_df=df[:1000]
train_input = train_df[:999]['CLEAN_PRICE']
train_output = train_df[1:]['CLEAN_PRICE']
print(len(train_input))
print(len(train_output))

test_df=df[1001:]
test_input = test_df[:673]['CLEAN_PRICE']
test_output = test_df[1:]['CLEAN_PRICE']
print(len(test_input))
print(len(test_output))

#normalize
train_input = train_input.values.tolist()
train_input = np.array(train_input)
train_input = train_input.reshape((len(train_input),1))
test_input = test_input.values.tolist()
test_input = np.array(test_input)
test_input = test_input.reshape((len(test_input),1))

train_output = train_output.values.tolist()
train_output = np.array(train_output)
test_output = test_output.values.tolist()
test_output = np.array(test_output)

train_input= norm(train_input)
test_input= norm(test_input)
train_output= norm(train_output)
test_output= norm(test_output)

print(train_input.shape, train_input.ndim)
print(train_input)
print(train_output)
print(test_input)
print(test_output)
```

```python
model = Sequential()
model.add(Dense(1,input_shape=(998,1), activation='relu'))
#model.add(Dense(32, activation='relu'))
#model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer=Adam(lr=0.95))
model.fit(train_input, train_output, epochs=500, verbose=1,batch_size=32)
```

```python
a = model.evaluate(test_input, test_output)
print(a)
b=model.predict(test_input)
print(b)
print(test_output)
```
